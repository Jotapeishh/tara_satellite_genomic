```python
import pandas as pd
import numpy as np

import os

from glob import glob
```

```python
input_dir = '../../01_data/01_biological_data'
file_name_md = 'metadata.tsv'

file_path = os.path.join(input_dir, file_name_md)

md = pd.read_csv(file_path, sep='\t', index_col=0)
md.head()
```

```python
indices_cflux = md['Mean Flux at 150m'].dropna().index.tolist()
indices_npp = md['NPP 8d VGPM (mgC/m2/day)'].dropna().index.tolist()
indices_fluorescence = md['Fluorescence'].dropna().index.tolist()

list_indices = {'Mean Flux at 150m' : indices_cflux,
                'NPP 8d VGPM (mgC/m2/day)' : indices_npp,
                'Fluorescence' : indices_fluorescence}
```

```python
input_sat_dir = '../../01_data/02_satellite_data_processed'

sat_data = sorted(glob(os.path.join(input_sat_dir, 'matrix_tara_world_adj_grids_*.tsv')))


intersections = {}
for file in sat_data:
    # Nombre del archivo como identificador
    file_name = os.path.basename(file)
    print(file_name)
    
    df = pd.read_csv(file, sep='\t', index_col=0)

    for key, indices in list_indices.items():
        intersection_indices = df.index.intersection(indices)
        col_name = f'{file_name},{key}'
        intersections[col_name] = intersection_indices.tolist()

max_length = max(len(indices) for indices in intersections.values())

results_df = pd.DataFrame({k: v + [None] * (max_length - len(v)) for k, v in intersections.items()})

# Guardar el DataFrame en un archivo .tsv
output_file = 'intersection_indices.tsv'
results_df.to_csv(output_file, sep='\t', index=False)
```
