```python
import pandas as pd
import netCDF4

from dateutil import parser
import os
```

```python
file_path = '../01_data/01_biological_data'
file_name = 'metadata_chile.tsv'

file = os.path.join(file_path,file_name)


md = pd.read_csv(file, sep='\t', index_col=0)
md_srf = md[md['Depth level'] == 'SRF']
md_srf
```

```python
from datetime import datetime
```

```python
def format_cleanup(datetime_str):
    date_str = datetime_str[:-6]
    old = date_str
    if "/" in date_str:
        date_format = "%m/%d/%Y"
    elif "-" in date_str:
        date_format = "%m-%d-%Y"
    if date_str[-3] == '-':
        date_str = date_str[:6] + "20" + date_str[6:]
    date = datetime.strptime(date_str, date_format)
    clean_date_str = datetime.strftime(date, '%Y%m')
    return clean_date_str
```

```python
md_srf_copy = md_srf.copy()
```

```python
md_srf_copy['date'] = md_srf_copy['datetime'].apply(format_cleanup)
```

```python
unique_dates = md_srf_copy['date'].unique()

unique_dates_df = pd.DataFrame(unique_dates, columns=['Unique_Dates_YYYYMM'])
unique_dates_df['Unique_Dates_YYYYMMDD'] = unique_dates_df['Unique_Dates_YYYYMM'] + '01'

print(unique_dates_df)


unique_dates_df['Unique_Dates_YYYYMM'].to_csv('unique_dates_yyyymm_ch.tsv', sep='\t', index=False)
unique_dates_df['Unique_Dates_YYYYMMDD'].to_csv('unique_dates_yyyymmdd_ch.tsv', sep='\t', index=False)

print("Archivo .tsv guardado exitosamente.")
```
